```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="sex", hue="sex", data=kashati)
```

```python
sns.countplot(x="survived",hue="sex",data=kashati)
```

```python
#scatter plot
plt.scatter(data=kashati, x="survived", y="who", color="sex")
```

```python
# scatter plot k liye hemsha dono numeric value honi chahiye mtlb(x axis and y axis)  aur dono hi apk pas numbers ho mtlb flot ya intger
# aur hue waly hisy may category honi chahiye
sns.scatterplot(data=kashati, x="fare", y="age", hue="pclass")
```

```python
sns.barplot(data=kashati, x="sex", y="fare", hue="pclass")
```

```python
sns.pairplot(kashati)
```

```python
sns.boxplot(data=kashati, x="sex", y ="fare", hue="pclass")
```

```python
sns.boxenplot(data=kashati, x= "sex", y="fare", hue="class")
```

```python
facetgrid = sns.FacetGrid(kashati, hue="survived", col="sex", 
                          margin_titles=True, palette={1: "seagreen",0:"grey"})
facetgrid.map(plt.scatter, "fare", "age", edgecolor="w").add_legend()
```

```python
sns.stripplot(data=kashati, x="sex", y ="fare", hue="class")
```

```python
#heatmap
corr = kashati.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("heat map of kashtai data")
```

```python
sns.violinplot(data=kashati, x="sex", y ="fare", hue="class")
```
