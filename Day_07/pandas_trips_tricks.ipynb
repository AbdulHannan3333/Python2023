# 16- splitting a string into multiple coloumn

```python
# import libraries
import pandas as pd
df = pd.DataFrame({
    'name':['javeria hannan', 'Hafsa Mueen', 'Aizey Abudallah'],
           'location':['Lahore, Pakistan', 'Faisalabad, Pakistan',
                        'karachi, Pakistan'] })
df
```

```python
df.name.str.split(' ').head()
```

```python
# split a cloumn into two coloumn
df.name.str.split(' ',expand=True)
```

```python
# add those spilt into two coloumn
df[['first_name','last_name']]= df.name.str.split(' ', expand=True)
df
```

```python
df[['city_name', 'country_name']]= df.location.str.split(', ',expand=True)
df
```

```python
# refine data manupulation
df[['first_name', 'last_name',	'city_name'	,'country_name']]
```
