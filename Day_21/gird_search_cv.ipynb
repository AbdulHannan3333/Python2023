```python
# Linear Regression 
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

X= df[['age']]
y= df['fare']
X.fillna(value=X.mean(), inplace=True)

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
# creating model
model = LinearRegression()

# define parameter gird
parm_gird = {'fit_intercept': [True,False]}
# object gird search cv(Creating the model)
gird_serach = GridSearchCV(model, parm_gird, cv=5, scoring='r2')

# train the mode
gird_serach.fit(X,y)

# print the best parameter
print('Best Parameter: ', gird_serach.best_params_)
```

# KNN Gird Search Cv

```python
#K Nearest Neighbour
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

X= df[['parch', 'pclass','sex','age','sibsp', 'fare']]
y= df['survived']
X.fillna(value=X.mean(), inplace=True)
X = pd.get_dummies(X, columns=['sex'])

from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import GridSearchCV
# creating model
model = KNeighborsClassifier()
# define parameter gird
parm_gird = {'n_neighbors':[ 5,7,15,5,2,5,256,4], 'weights': ['distance', 'uniform']}
# object gird search cv(Creating the model)
gird_serach = GridSearchCV(model, parm_gird, cv=5, scoring='f1')

# train the mode
gird_serach.fit(X,y)

# print the best parameter
print('Best Parameter: ', gird_serach.best_params_)
print('Best Score: ', gird_serach.best_score_)
```

```python
#K Nearest Neighbour
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset('titanic')

X= df[['parch', 'pclass','sex','age','sibsp', 'fare']]
y= df['survived']
X.fillna(value=X.mean(), inplace=True)
X = pd.get_dummies(X, columns=['sex'])

from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import GridSearchCV
# creating model
model = KNeighborsClassifier()
# define parameter gird 
#(0,40,1) iska mtlb  k 0 su ly kar 40 tak aur 2 k farq k sth
parm_gird = {'n_neighbors':np.arange(1,40,2), 'weights': ['distance', 'uniform']}
# object gird search cv(Creating the model)
gird_serach = GridSearchCV(model, parm_gird, cv=5, scoring='precision')

# train the mode
gird_serach.fit(X,y)

# print the best parameter
print('Best Parameter: ', gird_serach.best_params_)
print('Best Score: ', gird_serach.best_score_)
```

```python
#DTC
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset('titanic')

X= df[['parch', 'pclass','sex','age','sibsp', 'fare']]
y= df['survived']
X.fillna(value=X.mean(), inplace=True)
X = pd.get_dummies(X, columns=['sex'])

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
# creating model
model = DecisionTreeClassifier()
# define parameter gird 
#(0,40,1) iska mtlb  k 0 su ly kar 40 tak aur 2 k farq k sth
parm_gird = {'max_depth': [3,5,7, None], 'min_impurity_decrease':[2,3,4] }
# object gird search cv(Creating the model)
gird_serach = GridSearchCV(model, parm_gird, cv=5, scoring='precision')

# train the mode
gird_serach.fit(X,y)

# print the best parameter
print('Best Parameter: ', gird_serach.best_params_)
print('Best Score: ', gird_serach.best_score_)
```

```python
#RFC
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset('titanic')

X= df[['parch', 'pclass','sex','age','sibsp', 'fare']]
y= df['survived']
X.fillna(value=X.mean(), inplace=True)
X = pd.get_dummies(X, columns=['sex'])

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
# creating model
model = RandomForestClassifier()
# define parameter gird 
#(0,40,1) iska mtlb  k 0 su ly kar 40 tak aur 2 k farq k sth
parm_gird = {'n_estimators': [100,10,50],
            'max_depth': [3,5,7, None],
            'min_samples_split':[2,5,9],
            'max_features': ['sqrt','auto','log2']}
# object gird search cv(Creating the model)
gird_serach = GridSearchCV(model, parm_gird, cv=5, scoring='precision')

# train the mode
gird_serach.fit(X,y)

# print the best parameter
print('Best Parameter: ', gird_serach.best_params_)
print('Best Score: ', gird_serach.best_score_)
```
